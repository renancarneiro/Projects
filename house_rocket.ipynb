```python
import pandas as pd
dados = pd.read_csv('/content/datasets/kc_house_data.csv')
#funcao que converte de objcte (string) ->date
dados.head()

```

#3. Criar uma classificaÃ§Ã£o para o imÃ³veis, separando-os em baixo e alto padrÃ£o, de acordo com  preÃ§o. 


```python
#Acima de R$ 540.000 -> Alto PadrÃ£o (high_standard)
#Abaixo de R$ 540.000 -> Baixo PadrÃ£o (low_standard)
#dados[['id','price']].sort_values('price', ascending=False).head()
dados['level'] = 'standard'
dados.loc[dados['price'] > 540000, 'level'] = 'high_standard'
dados.loc[dados['price'] < 540000, 'level'] = 'low_standard'
dados.head()
```

#4. RelatÃ³rio ordenado pelo preÃ§o e contento as seguintes informaÃ§Ãµes: ( id do imÃ³vel, data que o imÃ³vel ficou disponÃ­vel para compra, o nÃºmero de quartos, o tamanho total to terreno, o preÃ§o, a classificaÃ§Ã£o do imÃ³vel ( alto e baixo padrÃ£o )

```python
report = dados[['id', 'date', 'bedrooms', 'sqft_lot','price', 'level']].sort_values('price', ascending=False)
report.to_csv('datasets/report.csv', index=False)
relatorio = pd.read_csv('datasets/report.csv')
relatorio.head()
```

#5. Mapa indicando onde as casas estÃ£o localizadas geograficamento.

```python
import plotly.express as px

data_mapa = dados[['id', 'lat', 'long', 'price']]

mapa = px.scatter_mapbox(data_mapa, lat='lat', lon='long', hover_name='id',
hover_data=['price'], color_discrete_sequence=['fuchsia'], zoom=8, height=300)

mapa.update_layout(mapbox_style='open-street-map')
mapa.update_layout(height=600, margin={'r':0, 't':0, 'l':0, 'b':0})
mapa.show()
mapa.write_html('datasets/mapa_houses.html')
```
